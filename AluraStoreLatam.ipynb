### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresos_tienda1 = tienda['Precio'].sum()
ingresos_tienda2 = tienda2['Precio'].sum()
ingresos_tienda3 = tienda3['Precio'].sum()
ingresos_tienda4 = tienda4['Precio'].sum()

print(f"Ingresos totales Tienda 1: {ingresos_tienda1}")
print(f"Ingresos totales Tienda 2: {ingresos_tienda2}")
print(f"Ingresos totales Tienda 3: {ingresos_tienda3}")
print(f"Ingresos totales Tienda 4: {ingresos_tienda4}")
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Precio'].sum()
print("Ventas por categoría Tienda 1:")
print(ventas_por_categoria_tienda1)

ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
print("\nVentas por categoría Tienda 2:")
print(ventas_por_categoria_tienda2)

ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
print("\nVentas por categoría Tienda 3:")
print(ventas_por_categoria_tienda3)

ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()
print("\nVentas por categoría Tienda 4:")
print(ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {calificacion_promedio_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {calificacion_promedio_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {calificacion_promedio_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {calificacion_promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
revenue_by_product_tienda = tienda.groupby('Producto')['Precio'].sum().sort_values(ascending=False)
print("Top 5 productos más vendidos en Tienda 1 (por ingresos):")
print(revenue_by_product_tienda.head(5))

revenue_by_product_tienda2 = tienda2.groupby('Producto')['Precio'].sum().sort_values(ascending=False)
print("\nTop 5 productos más vendidos en Tienda 2 (por ingresos):")
print(revenue_by_product_tienda2.head(5))

revenue_by_product_tienda3 = tienda3.groupby('Producto')['Precio'].sum().sort_values(ascending=False)
print("\nTop 5 productos más vendidos en Tienda 3 (por ingresos):")
print(revenue_by_product_tienda3.head(5))

revenue_by_product_tienda4 = tienda4.groupby('Producto')['Precio'].sum().sort_values(ascending=False)
print("\nTop 5 productos más vendidos en Tienda 4 (por ingresos):")
print(revenue_by_product_tienda4.head(5))
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio Tienda 1: {costo_envio_promedio_tienda1:.2f}")
print(f"Costo de envío promedio Tienda 2: {costo_envio_promedio_tienda2:.2f}")
print(f"Costo de envío promedio Tienda 3: {costo_envio_promedio_tienda3:.2f}")
print(f"Costo de envío promedio Tienda 4: {costo_envio_promedio_tienda4:.2f}")
```

```python
import numpy as np

# Prepare data for plotting
stores = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
average_ratings = [
    calificacion_promedio_tienda1,
    calificacion_promedio_tienda2,
    calificacion_promedio_tienda3,
    calificacion_promedio_tienda4
]
average_shipping_costs = [
    costo_envio_promedio_tienda1,
    costo_envio_promedio_tienda2,
    costo_envio_promedio_tienda3,
    costo_envio_promedio_tienda4
]

x = np.arange(len(stores))  # the label locations
width = 0.35  # the width of the bars

fig, ax1 = plt.subplots(figsize=(12, 7))

# Plot average ratings on the first y-axis
rects1 = ax1.bar(x - width/2, average_ratings, width, label='Calificación Promedio', color='lightskyblue')
ax1.set_xlabel('Tienda')
ax1.set_ylabel('Calificación Promedio', color='darkblue')
ax1.tick_params(axis='y', labelcolor='darkblue')

# Create a second y-axis for average shipping costs
ax2 = ax1.twinx()
rects2 = ax2.bar(x + width/2, average_shipping_costs, width, label='Costo de Envío Promedio', color='lightcoral')
ax2.set_ylabel('Costo de Envío Promedio', color='darkred')
ax2.tick_params(axis='y', labelcolor='darkred')

# Set title and x-axis ticks
plt.title('Calificación Promedio y Costo de Envío Promedio por Tienda')
ax1.set_xticks(x)
ax1.set_xticklabels(stores)

# Add legends
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))

plt.tight_layout()
plt.show()
```

Informe Ejecutivo: Análisis de Rendimiento y Estrategia de Venta de Tiendas
1. Introducción
El presente informe evalúa el rendimiento operativo y financiero de cuatro tiendas (Tienda 1, Tienda 2, Tienda 3 y Tienda 4) para proporcionar una base sólida en la toma de decisiones estratégicas. A través del análisis de métricas clave —como ingresos totales, ventas por categoría, calificaciones de clientes y costos de envío— se identifican las fortalezas y debilidades de cada sucursal para formular una recomendación justificada para el Sr. Juan sobre qué activos vender y cuáles conservar.

2. Resumen General de Hallazgos
Ingresos Totales: La Tienda 1 lidera la facturación con $1,150.8M, seguida de la Tienda 2 ($1,116.3M) y la Tienda 3 ($1,098.0M). La Tienda 4 registra los ingresos más bajos con $1,038.3M.

Satisfacción del Cliente: La Tienda 3 ostenta la calificación promedio más alta (4.05/5.00), lo que indica una excelente fidelidad y satisfacción. Por el contrario, la Tienda 1 presenta la calificación más baja (3.98/5.00).

Eficiencia Operativa (Costos de Envío): La Tienda 4 es la más eficiente logísticamente, con el costo de envío promedio más bajo ($23,459). La Tienda 1 es la más costosa en este rubro ($26,018).

Tendencias de Productos: Las categorías de Electrónicos y Electrodomésticos dominan las ventas en todas las tiendas. Productos como el TV LED UHD 4K, Refrigeradores y el iPhone 15 son los motores principales de ingresos a nivel general.

Patrones Geográficos: Las ventas y calificaciones se concentran fuertemente en las principales áreas metropolitanas de Colombia (Bogotá, Medellín, Cali, Cartagena), demostrando un mercado consolidado en zonas urbanas.

3. Análisis Detallado por Tienda (Fortalezas y Debilidades)
Tienda 1
Fortalezas: Posee la facturación más alta de las cuatro tiendas ($1,150,880,400) y un liderazgo absoluto en las categorías de mayor valor (Electrónicos y Electrodomésticos).

Debilidades: Presenta los costos de envío promedio más altos y la calificación de satisfacción del cliente más baja (3.98). Su alta dependencia de solo dos categorías la hace vulnerable a fluctuaciones del mercado, y sus ineficiencias logísticas podrían estar mermando la rentabilidad neta.

Tienda 2
Fortalezas: Muestra un rendimiento comercial muy sólido (segundo lugar en ingresos con $1,116,343,500). Tiene un excelente nivel de satisfacción (4.04) y un nicho diferenciador clave: es líder indiscutible en la venta de Instrumentos Musicales.

Debilidades: Sus costos de envío son el segundo más alto del grupo ($25,216), lo que representa un área de mejora en la competitividad logística.

Tienda 3
Fortalezas: Posee la mejor calificación promedio de clientes (4.05) y el segundo costo de envío más bajo ($24,805), lo que refleja una operación altamente eficiente y una clientela leal. Además, lidera significativamente en la categoría de Muebles.

Debilidades: Se ubica en el tercer lugar en volumen de ingresos totales ($1,098,019,600), lo que indica que aún tiene margen para escalar sus ventas generales y potenciar categorías secundarias.

Tienda 4
Fortalezas: Es la tienda logísticamente más eficiente, con el costo de envío más bajo de todas ($23,459). Mantiene una buena satisfacción del cliente (4.00) y destaca liderando la categoría de Juguetes.

Debilidades: Genera los ingresos totales más bajos ($1,038,375,700) y presenta el rendimiento más débil en la categoría de Electrodomésticos. Requeriría una inversión considerable para escalar su volumen de ventas al nivel de sus pares.

4. Conclusión y Recomendación Estratégica
Basado en el equilibrio entre rentabilidad, satisfacción del cliente y eficiencia operativa, se recomienda al Sr. Juan vender la Tienda 1 y la Tienda 4, y conservar la Tienda 2 y la Tienda 3.

Justificación de la Estrategia:
Desinversión Estratégica (Vender Tiendas 1 y 4):

Vender la Tienda 1 permite capitalizar su alto volumen de ingresos brutos en el mercado actual, deshaciéndose simultáneamente de la operación con los costos logísticos más altos y la menor satisfacción de clientes (riesgos a largo plazo).

Vender la Tienda 4 permite simplificar la cartera de negocios al eliminar la sucursal con menor facturación, evitando la inversión de capital y esfuerzo necesarios para escalar su volumen de ventas.

Enfoque en Valor y Eficiencia (Conservar Tiendas 2 y 3):

Al mantener estas dos sucursales, el Sr. Juan se queda con los negocios más equilibrados y saludables. Ambas tienen una facturación robusta, excelente retención y satisfacción de clientes (calificaciones > 4.04), y una oferta diversificada en categorías rentables (Instrumentos Musicales y Muebles). En particular, la Tienda 3 ofrece el perfil más atractivo para el crecimiento a largo plazo gracias a su combinación de alta lealtad de clientes y bajos costos operativos.
